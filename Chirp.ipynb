```python
import sys
import numpy as np, matplotlib.pyplot as plt
from func import State, Functions
np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(suppress=True, precision=4)
```

```python
state = State()
func = Functions(state)
```

```python
ch = func.chirp(alpha=0.0003, q=1)
```

```python
f = plt.figure(figsize=[16, 9])

plt.plot(state.r, ch, lw=4, c='black', label='t=0')
plt.xlim([state.r[1], state.r[-1]])
plt.grid()
plt.show()
```

```python
F = func.radial_function(ch, 0) 
```

```python
f = plt.figure(figsize=[16, 9])
plt.subplot(1, 2, 1)
plt.imshow(F)
plt.subplot(1, 2, 2)
plt.imshow(np.angle(F))
plt.show()
```
